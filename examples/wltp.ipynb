```python
import os
os.chdir(f"{globals()['_dh'][0]}/..")  # Set cwd up one level

from etron import e_tron
import pandas as pd
import numpy as np

from matplotlib import pyplot as plt
import seaborn as sns
sns.set()
```

```python
cycle = pd.read_csv("data/wltp-3.csv", sep="\t")
```

```python
ev = e_tron()
ev.simulate(cycle)
```

```python
fig = plt.figure(figsize=(10, 3))
ax = fig.add_subplot(111)

x = ev.log["Time (s)"]
y = ev.log["Velocity (mph)"]

ax.axvspan(xmin=0, xmax=589, alpha=0.1, color="green")
ax.axvspan(xmin=590, xmax=1022, alpha=0.1, color="yellow")
ax.axvspan(xmin=1023, xmax=1477, alpha=0.1, color="red")
ax.axvspan(xmin=1478, xmax=1800, alpha=0.2, color="red")

ax.set_title("WLTP 3B Drive Cycle", fontsize="large", fontweight="bold")
ax.set_xlabel("Time, sec", fontsize="large")
ax.set_ylabel("Velocity, mph", fontsize="large")

ax.plot(x, y, color="black")
```
