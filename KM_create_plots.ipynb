```python
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib
```

```python
# Read csv file into dataframe
df = pd.read_csv("DataSets/nyccol.csv")
```

```python
# Convert speed to kph
df["Velocity (Kph)"]=df["Velocity (mph)"]*0.6214
```

```python
plt.plot(df["Time (s)"],df["Velocity (mph)"])
plt.xlabel("Time (s)")
plt.ylabel("Velocity (mph)")
plt.show()
```
