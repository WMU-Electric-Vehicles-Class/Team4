```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
sns.set()
```

```python
data = {
    "Year of Release": [1881, 1882, 1888, 1895, 1896, 1921, 1960, 1966, 1967, 1968, 1988, 1996, 2008, 2021],
    "Top Speed (mph)": [7, 9, 8, 12, 12, 25, 30, 40, 40, 55, 60, 78, 125, 167.14],
}
df = pd.DataFrame(data)
```

```python
x = df["Year of Release"]
y = df["Top Speed (mph)"]

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, y, marker="o")

ax.set_title("EV Top Speed vs. Year of Release", size="large")
ax.set_xlabel(x.name)
ax.set_ylabel(y.name)
```
