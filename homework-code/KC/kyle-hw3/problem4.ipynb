```python
from matplotlib import pyplot as plt
from fastsim import vehicle, cycle, simdrive
```

```python
veh = vehicle.Vehicle(22)
veh.Scenario_name
```

```python
cyc = cycle.Cycle("hwfet")
```

```python
sim = simdrive.SimDriveClassic(cyc, veh)
sim.sim_drive()
```

```python
x = cyc.cycSecs
y1 = sim.soc * 100
y2 = cyc.cycMps * 2.23694

plt.rcParams["font.family"] = "Times New Roman"
fig = plt.figure(figsize=(10,4))
ax1 = fig.add_subplot(111)

ax1.plot(x, y1, color="C0")
ax1.set_title("Tesla S 60: SOC vs. Time", fontsize="large", fontweight="bold")
ax1.set_xlabel("Time [sec]", fontsize="large")
ax1.set_ylabel("SOC [%]", fontsize="large")
# ax.set_ylim(bottom=0, top=1)

fig.tight_layout()
```
