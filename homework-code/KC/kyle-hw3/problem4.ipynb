```python
from matplotlib import pyplot as plt
from fastsim import vehicle, cycle, simdrive
```

```python
veh = vehicle.Vehicle(22)
veh.Scenario_name
```

```python
cyc = cycle.Cycle("hwfet")
```

```python
sim = simdrive.SimDriveClassic(cyc, veh)
sim.sim_drive()
```

```python
x = cyc.cycSecs
y1 = sim.soc * 100
y2 = cyc.cycMps * 2.23694

fig = plt.figure(figsize=(10,8))
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212, sharex=ax1)

ax1.plot(x, y1, color="C0")
ax1.set_title("SOC vs. Time")
ax1.set_xlabel("Time [sec]")
ax1.set_ylabel("SOC [%]", fontsize="large")
# ax.set_ylim(bottom=0, top=1)

ax2.plot(x, y2, color="C1")
ax2.set_xlabel("Time [sec]")
ax2.set_ylabel("Velocity [mph]")
ax2.set_title("Velocity vs. Time", fontsize="large")

fig.tight_layout()
```
