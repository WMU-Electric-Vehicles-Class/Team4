```python
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib
import numpy as np 
```

```python
# Read csv file into dataframe
df = pd.read_csv("DataSets/nyccol.csv")
```

```python
# Convert speed to kph
df["Velocity (kph)"]=df["Velocity (mph)"]*0.6214
```

```python
# Convert speed to acceleration
df["Acceleration (m/s^2)"]=df["Velocity (mph)"]*0.44704
```

```python
# Convert mph to mps
df["Velocity (mps)"]=df["Velocity (mph)"]/3600
```

```python
# calculation of distance
df["Distance (mi)"]=np.cumsum(df["Velocity (mps)"])
```

```python
plt.plot(df["Distance (mi)"],df["Velocity (mph)"])
plt.xlabel("Distance (mi)")
plt.ylabel("Velocity (mph)")
plt.show()
```

```python
plt.plot(df["Time (s)"],df["Velocity (kph)"])
plt.xlabel("Time (s)")
plt.ylabel("Velocity (kph)")
plt.show()
```

```python

plt.plot(df["Time (s)"],df["Acceleration (m/s^2)"])
plt.xlabel("Time (s)")
plt.ylabel("Acceleration (m/s^2)")
plt.show()
```
