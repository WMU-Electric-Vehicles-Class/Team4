```python
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib
import numpy as np 
```

```python
# Read csv file into dataframe
df = pd.read_csv("DataSets/sc03col.csv")
```

```python
# Convert speed to kph
df["Velocity (kph)"]=df["Vehicle Speed"]*0.6214
```

```python
# Convert speed to acceleration
df["Acceleration (m/s^2)"]=df["Vehicle Speed"]*0.44704
```

```python
# Convert mph to mps
df["Velocity (mps)"]=df["Vehicle Speed"]/3600
```

```python
# calculation of distance
df["Distance (mi)"]=np.cumsum(df["Velocity (mps)"])
```

```python
plt.plot(df["Distance (mi)"],df["Vehicle Speed"])
plt.xlabel("Distance (mi)")
plt.ylabel("Velocity (mph)")
plt.show()

plt.plot(df["Test Time"],df["Velocity (kph)"])
plt.xlabel("Time (s)")
plt.ylabel("Velocity (kph)")
plt.show()

plt.plot(df["Test Time"],df["Acceleration (m/s^2)"])
plt.xlabel("Time (s)")
plt.ylabel("Acceleration (m/s^2)")
plt.show()
```
